## 1. История развития ОС Linux. 

[Интервью с Линусом Торвальдсом (начиная со 2-ой главы)](https://github.com/limkokhole/just-for-fun-linus-torvalds/blob/master/just-for-fun.pdf)

1. 1991 — первый релиз в Usenet. 

Прототипом для будущего ядра стала операционная система **MINIX**. MINIX был создан Эндрю Таненбаумом в качестве учебной операционной системы, демонстрирующей архитектуру и возможности UNIX. Название Linux проект получил не сразу. Изначально он назывался **FreaX**. Совместимость с UNIX в этот момент означала, что операционная система должна поддерживать стандарт **POSIX**. 

>**POSIX** — это функциональная модель совместимой с UNIX операционной системы, в которой описано, как должна вести себя система в той или иной ситуации, но не приводится никаких указаний, как это следует реализовать программными средствами. POSIX описывал те свойства UNIX-совместимых систем, которые были общими для разных реализаций UNIX на момент создания этого стандарта. В частности, в POSIX описаны системные вызовы, которые должна обрабатывать операционная система, совместимая с этим стандартом.

На самых ранних стадиях Линус Торвальдс обсуждал свою работу и возникающие трудности с другими разработчиками в телеконференции comp.os.minix в сети [**Usenet**](https://ru.wikipedia.org/wiki/Usenet), посвящённой операционной системе MINIX. Ключевым решением Линуса стала публикация исходных текстов ещё малоработоспособной первой версии ядра под свободной лицензией [**GNU GPL**](https://ru.wikipedia.org/wiki/GNU_General_Public_License). Благодаря этому многие получили возможность самостоятельно компилировать и тестировать это ядро, участвовать в обсуждении и исправлении ошибок, а также присылать исправления и дополнения к исходным текстам Линуса. Теперь над ядром работал уже не один человек, разработка пошла быстрее и эффективнее. 

Поворотным на данном этапе стал момент, когда ОС смогла запустить **компилятор С.**

Другим отличием ОС от существовавших тогда альтернатив была **архитектура, изначально ориентированная на сетевое взаимодействие.**

2. 1992 — поддержка командной оболочки, компилятора файловых систем

3. 1993 — поддержка пользовательского окружения GNU

4. 1994 — релиз 1.0. Графический интерфейс.
 
5. 1996 — релиз 2.0. Поддержка SMP (параллельной работы нескольких процессоров) + Red Hat

6. 1998 — поддержка в промышленных продуктах крупных компаний (Oracle, IBM).

7. 1998 - 2012 — вытеснение коммерческих Unix (HP-UX, Solaris, AIX и т. п.) с серверного рынка.

8. 2016 — windows subsystem for linux.

[Хронология более подробно](https://ru.wikipedia.org/wiki/%D0%AF%D0%B4%D1%80%D0%BE_Linux#%D0%A5%D1%80%D0%BE%D0%BD%D0%BE%D0%BB%D0%BE%D0%B3%D0%B8%D1%8F)

## 2. Ядра и их версии. Ванильные ядра.

Ядро Linux разрабатывается под эгидой Linux Foundation. Она выкладывает отчеты по разработке ядра примерно раз в год;
в ней участвуют как волонтеры, так и компании (Intel, Redhat, Samsung, IBM, Google, Huawei, AMD и т.д.). В дополнение к официальным (ванильным) версиям ядра существуют альтернативные ветки, которые могут быть взяты из различных источников. Как правило, разработчики дистрибутивов Linux поддерживают свои собственные версии ядра, например, включая в них драйверы устройств, которые ещё не включены в официальную версию. Например, в ядре 3.10.0-1062.7.1 из дистрибутива RHEL 7 присутствует функционал из версий ядра вплоть до 4.18.

Общее руководство осуществляют Линус Торвальдс и Грег Кроа-Хартман. В рамках отдельных подсистем ядра решения о принятии тех или иных изменений принимают мейнтейнеры.

Релизы ядра доступны на https://www.kernel.org/. Рекомендуется скачать архив с последней версией ядра, распаковать, ознакомиться со структурой.

Цикл разработки ядра состоит из трех фаз:

- **merge window**: в master-ветку ядра принимаются изменения, подготовленные разработчиками и одобренные мейнтейнерами (см. файл MAINTAINERS в архиве ядра). Типичная длительность — 2 недели.

- **заморозка изменений**, с данного момента в текущую версию ядра принимаются только фиксы багов. выпуск релиз-кандидатов. Обычно их 6—8, типичное наименование - `<следующая версия>-rc<номер релиз-кандидата>` (например, `5.5-rc4`). Частота выпуска релиз-кандидатов — 1 неделя. В момент, когда руководитель, выпускающий данную версию (Линус или Грег) решает, что в текущем релиз-кандидате багов мало, происходит релиз.

- **релиз очередной версии ядра** (сопровождающийся открытием merge window для следующей версии). По мере нахождения багов выпускаются исправленные версии — например, 5.4.1, 5.4.2 и т. п. Длительность цикла поддержки большинства ядер — 3 месяца; периодически выпускаются LTS-версии (long term support) ядра, поддерживаемые 2 года.

### [Нумерация версий](https://ru.wikipedia.org/wiki/%D0%AF%D0%B4%D1%80%D0%BE_Linux#%D0%9D%D1%83%D0%BC%D0%B5%D1%80%D0%B0%D1%86%D0%B8%D1%8F_%D0%B2%D0%B5%D1%80%D1%81%D0%B8%D0%B9_%D0%B4%D0%BE_30_%D0%BC%D0%B0%D1%8F_2011)

Изначально смена первого числа версии ядра (1.0, 2.0) была приурочена к глобальной переработке ядра; второе число было четным для стабильных веток (2.0, 2.2, 2.4) и нечетной для активно разрабатываемой (2.1, 2.3, 2.5). Начиная с ядер ветки 2.6 в 2004 году цикл разработки пришел к современному виду: очередная версия получала номер 2.6.X. Вместо версии 2.6.40 было выпущено ядро 3.0 без каких-либо кардинальных изменений; очередные версии получили номера 3.Х. Выпуск версий 4.0 и 5.0 также не был связан с крупными изменениями ядра.

## 3. Дистрибутивы ОС Linux. Лицензия GPL и свободы которые она дает.

Все дистрибутивы основаны на едином семействе ядер, однако набор служебных программ, дополняющих ядро, может существенно варьироваться. Дистрибутивы различаются по своему назначению, наличию служб поддержки и степени популярности.

Одни дистрибутивы создавались в расчете на промышленное применение, другие — нет. Какой-нибудь дополнительный программный компонент системы промышленного уровня, на первый взгляд малозначительный, в действительности существенно упрощает администрирование.

Принимая решение в пользу того или иного дистрибутива, обращайте внимание не только на его функциональные возможности, но и на то, как будет протекать сотрудничество вашей организации и поставщика дистрибутива в ближайшие годы.
Задайте себе ряд важных вопросов.
- Будет ли дистрибутив существовать в ближайшие пять лет?
- Будут ли оперативно устраняться бреши в системе защиты?
- Будут ли оперативно выпускаться новые версии программных продуктов?
- Будет ли оказана помощь в решении возникших проблем?

Обновляемый перечень дистрибутивов можно найти по следующим адресам: www.lwn.net/Distributions и
www.distrowatch.com

### Основные отличия между дистрибутивами

1. Cпособ распространения ПО, входящего в основные репозитории (в исходных кодах / бинарных пакетах)
2. Политике обновления ПО (постоянное обновление до последних мажорных версий / обновление в пределах мажорной версии / сохранение совместимости и выборочное бекпортирование функционала из новых версий)
3. Cрок поддержки релизов.

Основные семейства — Debian, Red Hat Enterprise Linux / Fedora, Suse, Gentoo, Arch Linux.

Первым дистрибутивом в современном понимании, получившим широкое распространение, стал **Slackware**, созданный Патриком Фолькердингом. Он был широко известен пользователям Linux уже к **1994** году. 

### Debian

Продукт принципиально некоммерческий, отличающийся высокой стабильностью. Появился в 1993 году. Стабильные и тестируемая версии операционной системы Debian называются именами персонажей мультфильма «История игрушек». Нестабильная версия дистрибутива Debian постоянно носит кодовое имя Сид, по имени отрицательного персонажа из мультфильма, который ломал игрушки. 

Именно в Debian появился менеджер пакетов (APT). В настоящее время Debian — самый популярный дистрибутив Linux среди пользователей, являющихся профессионалами в области безопасности информационных технологий, так как имеет самую большую скорость реагирования на обнаруженные угрозы. Он стал основой семейства дистрибутивов, в которые входит **Mint**, **Tails** и **Ubuntu**.

**Ubuntu** основан на дистрибутиве Debian и использует его систему пакетирования.
Ubuntu в настоящее время финансируется за счет благотворительных пожертвований южноафриканского предпринимателя Мака Шаттлворта (Mark Shuttleworth). Он выпускается в двух основных вариантах: Desktop Edition и Server Edition. В сущности, они идентичны, но ядро Server Edition предварительно настроено для использования на сервере и не инсталлирует графический интерфейс (GUI) или такие GUI-приложения, как OpenOffice.

### Red Hat

Red Hat - это компания, занимающаяся продажей и поддержкой дистрибутивов. Появилась в 1995 году. При этом стоимость продукта складывается из рабочего времени программистов, собирающих и/или пишущих ПО, цены носителей, маркетинга, логистики и службы поддержки.  В числе услуг, оказываемых компанией, техническая поддержка, обучение системных администраторов и разработчиков, приём практических лабораторных экзаменов (таких как **RHCT** и **RHCE**). Red Hat занимается распространением дистрибутивов **Red Hat Enterprise Linux** (распространяется по годовой подписке) и **Fedora** (распространяется свободно). 

Сначала эти дистрибутивы были схожими, но за последние пять лет Fedora претерпела значительные изменения, и теперь эти две системы не синхронизированы. RHEL отличается высоким уровнем поддержки и стабильностью, но ее по существу невозможно использовать, не приобретя лицензию в компании Red Hat.

Проект **CentOS** (centos.org) концентрирует исходный код, который компания Red Hat обязана распространять в соответствии с различными лицензионными соглашениями (наиболее значимой является общедоступная лицензия GNU), и комплектует их в
законченный дистрибутив, который во многом подобен дистрибутиву Red Hat Enterprise Linux, но доступен бесплатно. Этот дистрибутив лишен торговой марки компании Red Hat и не содержит некоторых запатентованных программных средств, но в остальном аналогичен платной версии. CentOS стремится к полной совместимости с RHEL, как по двоичному коду, так и по степени исправления ошибок.

### Suse

Компания SUSE, которая теперь является одним из подразделений корпорации Novell, пошла по пути Red Hat и начала распространять два связанных дистрибутива: **openSUSE**, который содержит только бесплатное программное обеспечение, и платный **SUSE Linux Enterprise**, который включает средства формальной поддержки и предоставляет несколько дополнительных возможностей. 


### Gentoo

Ориентирован на оптимизацию под конкретную машину, таким образом распространяется в виде исходников и предоставляет возможность пользователю компилировать их самостоятельно, таким образом систему можно запустить в том числе на эзотерических процессорных архитектурах

### Arch

Первый релиз 2002 года. В официальных репозиториях Arch Linux содержится преимущественно свободное ПО, однако дистрибутив не одобрен FSF, так как Arch придерживается лояльной политики в отношении несвободного ПО. Имеет собственный пакетный менеджер pacman, пакеты преимущественно бинарные под х86-64, кроме того все пакеты из AUR собираются на машине пользователя


## GNU GPL

Это [лицензия](https://ru.wikipedia.org/wiki/GNU_General_Public_License#%D0%A1%D0%B2%D0%BE%D0%B1%D0%BE%D0%B4%D1%8B_%D0%B8_%D0%BE%D0%B1%D1%8F%D0%B7%D0%B0%D1%82%D0%B5%D0%BB%D1%8C%D1%81%D1%82%D0%B2%D0%B0) на свободное программное обеспечение, созданная в рамках проекта GNU в 1988 г., по которой автор передаёт программное обеспечение в общественную собственность. Вторая версия этой лицензии была выпущена в 1991 году, третья версия, после многолетней работы и длительной дискуссии — в 2007 году. 

Лицензия дает свободу:

- Запускать
- Пользоваться
- Модифицировать
- Продавать

а также гарантировать, что и пользователи всех производных программ получат вышеперечисленные права. Принцип «наследования» прав называется «копилефт» (транслитерация c англ. copyleft) и был придуман Ричардом Столлманом. 

## 4. [История](https://www.redhat.com/en/blog/rhelvolution-brief-history-red-hat-enterprise-linux-releases-early-days-rhel-5) коммерческого успеха дистрибутива Red Hat

Изначально компания разрабатывала единый дистрибутив Red Hat Linux. В 2002 году появился **RHEL2.1**. Он предоставлял инсталлер, компилятор и драйвера для поддержки различного оборудования. 

В 2003 году **RHEL3** начал конкурировать с системами Unix на **промышленных рынках**, уйдя в сторону от рабочих станций и узкоспециализированных задач типа принтерных и файловых серверов. В это время процессоры Intel становились серьезной конкуренцией специализированным чипам, поддерживающим проприетарные предложения UNIX. Делался уклон в сторону производительности и масштабируемости. Результатом этой работы стало то, что RHEL начал устанавливать новые мировые рекорды в производительности баз данных TPC, помогая продвинуть роль RHEL в центре обработки данных от «маленьких коробок в углу» к «рабочим нагрузкам предприятия». Это действительно начало расшатывать укоренившуюся позицию проприетарного Unix на проприетарном оборудовании. Начало конца «войн Unix» позднее вытеснило совместная мощь сообщества открытого исходного кода на все более мощные товарные процессоры Intel.

**RHEL4** (2005). Волна клиентов привела к появлению нового набора требований для работы в расширяющемся пространстве имен в Интернете - главным образом, безопасности. Чтобы удовлетворить потребности в безопасности, инженеры Red Hat тесно сотрудничали с сообществом, правительственными учреждениями и финансовыми корпорациями в создании защитной оболочки **SELinux** и средств контроля доступа.

Прим: SELinux и KVM являются модулями ядра.

>**SELinux** — это реализация системы принудительного управления доступом доступа(mandatory access control — MAC), в которой все привилегии назначаются администраторами. В среде MAC пользователи не могут делегировать свои права и не могут устанавливать параметры контроля доступа на объекты, которыми они (пользователи) владеют. SELinux (как проект) был разработан Агентством национальной безопасности США (АНБ), но с конца 2000 года был передан разработчикам открытого кода.

Внедрение SELinux было огромной проблемой интеграции, так как требовало изменений кода буквально в сотнях компонентов, и его успех можно напрямую отнести к разработчикам Red Hat, которые внесли основной вклад во все эти компоненты. Это было ухабистое путешествие через несколько выпусков, чтобы привести SELinux к более широкому распространению.

В 2007 году компьютеры становились все больше и больше, и для эффективного использования часто требовалась консолидация рабочей нагрузки, главным образом посредством виртуализации. Ранее виртуализация была доступна только через проприетарных поставщиков по высоким ценам. Первоначальная реализация виртуализации в Red Hat Enterprise Linux называлась Xen, которая впоследствии была заменена виртуальной машиной Linux на основе ядра (**KVM**). 

Управляемый Red Hat (и многими другими), KVM эволюционировал для реализации виртуальных машин как процессов. Этот простой и элегантный дизайн позволил многократно использовать многие уже существующие компоненты Linux, что послужило началом мощной эволюции в пространстве виртуализации. Это еще раз подтолкнуло Linux (и открытый исходный код в целом) к мировым рекордным показателям производительности и масштаба, демонстрируя мощь открытого исходного кода и RHEL в продвижении ведущих инноваций.

То, что было уникальным для RHEL (в дополнение к лидерским показателям), основывалось на возможностях безопасности SELinux, что привело к сильной изоляции гостевой виртуальной машины, что требуется для сертификации рабочей нагрузки. Эта комбинация функций (безопасность + виртуализация) в одном продукте демонстрирует мощные возможности Linux.

Например, постоянные улучшения масштабируемости при использовании больших систем ЦП и памяти были немедленно использованы в реализации виртуализации. Это позволило клиентам легче добиваться лучшего использования своего оборудования, выполняя существующие рабочие нагрузки в виртуализированных системах.

Что касается виртуализации, то мы не только поддерживали гостей RHEL, но позже поддерживали гостей Microsoft Windows. Поддержка гостевых систем Windows стала началом все более тесных отношений между Red Hat и Microsoft, которые продолжаются в **RHEL 8** в форме оптимизаций Azure.

## 5. Список поддерживаемых файловых систем ОС Linux. LVM

Основным назначением файловой системы является упорядочение хранимых ресурсов системы (т.е. файлов). Отсутствие файловых систем сначала было преимуществом, так как позволяло экономить ресурсы. Файловую систему можно представить состоящей из четырех основных компонентов:
- пространство имен — методы именования объектов и организации их в виде единой иерархии;
- API — набор системных вызовов для перемещения между объектами и управления ими;
- модель безопасности — схема защиты, сокрытия и совместного использования объектов;
- реализация — программный код, который связывает логические модели с дисковой подсистемой

В современных ядрах систем определен абстрактный интерфейс, позволяющий работать с различными файловыми системами. Одни части файлового дерева обрабатываются традиционной дисковой подсистемой, другие — управляются специальными драйверами ядра. Например, за работу файловых систем — как сетевых (Network File System — NFS), так и общих межсетевых (Common Internet File System — CIFS), — отвечает драйвер, который перенаправляет запросы серверу, расположенному на другом компьютере.

Файловое дерево формируется из отдельных частей, называемых файловыми системами, каждая из которых содержит корневой каталог и список его подкаталогов и файлов. Термин “файловая система” имеет, по сути, два значения. С одной стороны, это составная часть файлового дерева, а с другой — все файловое дерево и алгоритмы, с помощью которых ядро управляет им. Как правило, значение термина становится ясным из контекста.

## Поддерживаемые файловые системы

Linux поддерживает больше файловых систем, чем любая другая разновидность UNIX.

- [Журналируемые](https://ru.wikipedia.org/wiki/%D0%96%D1%83%D1%80%D0%BD%D0%B0%D0%BB%D0%B8%D1%80%D1%83%D0%B5%D0%BC%D0%B0%D1%8F_%D1%84%D0%B0%D0%B9%D0%BB%D0%BE%D0%B2%D0%B0%D1%8F_%D1%81%D0%B8%D1%81%D1%82%D0%B5%D0%BC%D0%B0) (сначала запись идет в журнал, затем на диск, если операция не проведена успешна, попытка повторяется)
> Операция записи может быть прервана например при физической неисправности или отключении устройства
- Нежурналируемые

### Семейство Ext

>**ext** \- 1992
>
>**ext2** \- 2001. Нежурналируемая файловая система. Ограничения: размер файла 2 Тб, размер тома 32 Тб. Неэффективная работа с директориями с большим количеством файлов.
>
>**ext3** \- 2002. Журналируемая ФС с поддержкой htree для более эффективной работы с директориями, содержащими большое кол-во файлов. Не нуждается в дефрагментации
>
>**ext4** \- 2006-2008. Ограничение размера ФС — 1 экзобайт, размера файла — 32 Тб. Поддержка преаллокации, отложенной аллокации, [extent-ов](https://en.wikipedia.org/wiki/Extent_(file_systems)). В два раза больше и эффективнее, поскольку с ней происходит переход на 64-битные фс.


### Другие FS

- xfs — журналируемая ФС с акцентом на параллелизм и поддержку эффективной работы с большими файлами. Ограничение размера — 8 экзобайт.

- btrfs позиционируется как ФС нового поколения. Поддерживает copy on write, прозрачное сжатие, снепшоты, объединение устройств в raid-массивы. Максимальный размер — 8 экзобайт.

- множество файловых систем для использования на определенных типах устройств, разработанных для специальных задач и т. п. (f2fs, overlayfs).

## LVM

**Logical Volume Manager** — менеджер логических томов. Является модулем ядра. В отличие от разделов жёсткого диска, размеры логических томов можно легко менять. 

Устройство LVM:

- **Физический том(англ. physical volume)**: Раздел на жёстком диске или жёсткий диск на котором создаются логические тома.
- **Группа томов (англ. volume group)**: Набор физических томов в один объект. Содержат логические тома. Группу томов можно представить как жёсткие диски.
- **Логический том (англ. logical volume)**: аналогичен разделу (hda1, sdа1) на не-LVM системах. Так же, как и на них, представляется как блочное устройство и может нести файловую систему.
- **Физический диапазон (англ. physical extent)** : Небольшая часть диска(обычно имеет размер 4MB), которая может быть добавлена к логическому тому. Может быть добавлен к любому разделу.
- **Логические диапазоны (англ. logical extent)**: диапазоны, на которые разбивается логический том. Их объём одинаков по всей группе томов.


Преимущества LVM:

- использование любого количества жёстких дисков или их разделов как один большой раздел.
- логические тома могут быть "размазаны" на несколько жёстких дисков. Максимальный размер равен объёму всех жёстких дисков.
- изменение/создание/удаление логических томов в любом виде. То есть не зависит от физического расположения тома.
- снапшоты позволяют создавать резервную копию замороженной файловой системы практически на лету.
- поддержка устройств для различных целей, включая прозрачное шифрование файловой системы и кэширования часто используемых данных.
- динамическое увеличение размера логических томов по мере их наполнения. 

[Статья на ArchWiki](https://wiki.archlinux.org/index.php/LVM_(%D0%A0%D1%83%D1%81%D1%81%D0%BA%D0%B8%D0%B9)#%D0%A1%D0%BE%D0%B7%D0%B4%D0%B0%D0%BD%D0%B8%D0%B5_%D1%84%D0%B8%D0%B7%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B8%D1%85_%D1%82%D0%BE%D0%BC%D0%BE%D0%B2(PV))

## 6. Стандарт POSIX и его влияние на ОС Linux. FHS

### [POSIX](https://ru.wikipedia.org/wiki/POSIX) / ISO/IEC/IEEE 9945:2009

Это набор стандартов, описывающих интерфейсы между операционной системой и прикладной программой (системный API), библиотеку языка C и набор приложений и их интерфейсов. Стандарт создан для обеспечения совместимости различных UNIX-подобных операционных систем и переносимости прикладных программ на уровне исходного кода. Поддержка стандарта POSIX в линуксе позволила портировать многие существующие приложения с минимальными изменениями, что способствовало его распространению. Название «POSIX» было предложено Ричардом Столлманом.

### [FHS](https://ru.wikipedia.org/wiki/FHS)

Это стандарт, унифицирующий местонахождение файлов и каталогов с общим назначением в файловой системе UNIX.

В отличие от ОС Windows, структура директорий в которой привязана к присутствующим в системе разделам (c:, d:) и т. п., в UNIX-подобных ОС структура директорий не зависит от конкретной конфигурации устройств хранения: существует единая иерархия дерева директорий, а разнообразные хранилища (диски, сетевые ФС, псевдо-ФС и т. п.) могут монтироваться к произвольным путям в вышеупомянутом дереве. Каталог самого верхнего уровня называется корневым (root).

Информацию об иерархии можно получить командой `man hier`.

- `/bin` - Бинарные системные утилиты
- `/sbin` - Основные утилиты типа init и iptables
- `/media` - Точки монтирования съемных носителей
- `/var` - Изменяемые файлы, такие как файлы регистрации, временные почтовые файлы
- `/tmp` - Временные файлы (после перезагрузки удаляются)
- `/boot` - Загрузочные файлы (загрузчик, ядро)
- `/dev` - Файлы устройств
- `/home` - Домашние каталоги пользователей 
- `/root` - Домашний каталог суперпользователя
- `/proc` - Виртуальная файловая система, представляющая состояние ядра операционной системы и запущенных процессов в виде файлов

Одни системы хранят совместно используемые библиотечные файлы и прочие важные программы (например, препроцессор языка С) в каталоге `/lib`. Другие переместили эти элементы в каталог `/usr/lib`, оставив для каталога `/lib` роль символьной ссылки.

Огромное значение имеют также каталоги `/usr` и `/var`. В первом хранится большинство стандартных программ и другие полезные компоненты, в частности интерактивная документация и библиотеки. Совсем не обязательно, чтобы каталог `/usr` был отдельной файловой системой, однако для удобства администрирования его, как правило, монтируют именно так. 
В каталоге `/var` содержатся буферные каталоги, журнальные файлы, учетная информация и прочие компоненты, специфичные для каждого компьютера. Поскольку при возникновении проблем журнальные файлы быстро разрастаются, рекомендуется помещать каталог `/var` в отдельную файловую систему.

## 7. Работа с устройствами как с файлами. Процесс подключения устройств к системе FHS.

Взаимодействие с многими аппаратными устройствами в UNIX-подобных ОС осуществляется через специальные файлы, размещенные на псевдо-фс **devtmpfs**.

Примерами подобных устройств являются HDD/SSD (типичные названия устройств sdX / nvmeX), терминалы (ttyX), псевдотерминалы (ptsX), специальные устройства null (принимает на запись произвольные данные и забывает про них; при чтении отдает EOF), zero (при чтении отдает произвольное количество нулей), random и urandom (источники произвольного количества случайных чисел с гарантией использования достаточного кол-ва [энтропии](https://en.wikipedia.org/wiki/Entropy_(computing)) и без гарантии;).

Псевдо-файловые устройства делятся на **блочные** и **символьные**. Операции ввода/вывода с символьными устройствами осуществляются **побайтово**, с блочными — **поблочно** (размер блока зависит от свойств конкрентного устройства).

**Блоком** называется группа объединенных между собой секторов. Между блоками записывается _суперблок_ (или несколько, в зависимости от жд).

Помимо блоков в файловой системе хранятся блоки хранения метаданных _i-node_.

>**Индексный дескриптор (айнода)** — это структура данных в традиционных для ОС UNIX файловых системах (ФС), таких как UFS, ext4. В этой структуре хранится метаинформация о стандартных файлах, каталогах или других объектах файловой системы, кроме непосредственно данных и имени. Cм `man inode`

Индексные дескрипторы хранят информацию о файлах, такую как принадлежность владельцу (пользователю и группе), режим доступа (чтение, запись, запуск на выполнение) и тип файла. Существует определённое число индексных дескрипторов, которое указывает максимальное количество файлов, допускаемое определённой файловой системой. Обычно при создании файловой системы примерно 1 % её объёма выделяется под индексные дескрипторы. 

Базовая карта индексных дескрипторов хранится в суперблоке. Там же размещается таблица индексных дескрипторов, а после нее идет уже блок данных.

```
        --------------
        | суперблок  |
        --------------
        | опис.груп. |
        --------------
        |block bitmap|
        --------------
        |inode bitmap|
        --------------
        | inode table|
        --------------
        |    data    |
        --------------
```


Создание файла устройства из командной строки осуществляется с помощью команды mknod, из кода программы — с помощью библиотечной функции **mknod** (http://man7.org/linux/man-pages/man1/mknod.1.html и https://linux.die.net/man/3/mknod).

Процесс «подключения устройств» к дереву директорий называется **монтированием**. Участниками операции монтирования являются файловая система, находящаяся на блочном устройстве, удаленном хосте, в файле и т. п., а также точка монтирования — уже существующая директория, внутри которой будет доступно содержимое монтируемой ФС. Каталог в своем сегменте данных и/или инлайновом сегменте метаданных хранит информацию о других нодах, размещенных внутри данной директории.

Аргумент `-t` следует указывать, если автоматическое распознавание типа ФС по тем или иным причинам невозможно или не приводит к желаемому результату.

Аргумент `-o` позволяет указать конкретные опции монтирования. Набор опций, поддерживаемых той или иной ФС, приведен на соответствующей странице руководства — например, man ext4, раздел MOUNT OPTIONS.

Примеры:

- `mount /dev/sdb1 /mnt` — монтирование файловой системы, находящейся на разделе `/dev/sdb1` в директорию `/mnt`

- `mount /dev/sdb1 -t ext4 -o noatime /mnt` — то же самое, но с указанием типа ФС и опции `noatime`
- `mount 192.168.1.1:/path/to/websites /var/www` — монтирование сетевой файловой системы (NFS), находящейся на хосте 192.168.1.1 по пути `/path/to/websites` в директорию `/var/www`
- `mount -o loop /home/vasya/mycoolgame.iso /mnt` — монтирование образа ФС, находящегося в файле `/home/vasya/mycoolgame.iso` в директорию `/mnt`

При перезагрузке монтирование, выполненное вышеуказанным способом, не сохраняется. Список ФС, подлежащих монтированию при загрузке, приводится в [`/etc/fstab`](http://man7.org/linux/man-pages/man5/fstab.5.html)

Команда `umount -f `предназначена для “насильственного” демонтирования занятой файловой системы и поддерживается во всех рассмотренных нами системах. Вместо использования команды `umount -f`, когда оказывается, что файловая система, которую вы пытаетесь демонтировать, занята, запустите команду `fuser`, чтобы узнать, какие процессы работают с файловой системой. Команда `fuser -с точка_монтирования` выводит идентификаторы всех процессов, обращающихся к файлам или каталогам указанной файловой системы, а также ряд буквенных кодов, которые отображают природу этой активности.

## 8. CLI ОС Linux. Основные команды навигации.

В качестве командной оболочки в linux чаще всего используется bash. Примеры навигации в bash:

- `pwd`
- `ls`.
    Основные опции:

    - l — отображение свойств файлов и директорий;

    - d — отображение свойств указанной директории, а не ее содержимого;

    - t — сортировка отображаемых файлов по времени изменения от новых к старым;

    - a — включение отображения скрытых файлов и директорий (скрытыми являются файлы и директории, название которых начинается с точки - '.').
- `cd`. Без аргументов — перейти в домашнюю директорию текущего пользователя.

Основные разновидности путей — **относительный** и **абсолютный**. Абсолютный путь начинается с `/` - корневой директории — и не зависит от того, в какой директории он применяется. Относительный путь не начинается с / и применяется относительно текущей директории.

Тильда (~) в пути обозначает домашнюю директорию текущего пользователя; ~<логин> - домашнюю директорию пользователя, логин которого приведен; например, ~vasya — домашняя директория пользователя vasya.

## 9. Shells ОС Linux. Процесс обработки введенных команд.

Работа в командной строке осуществляется при помощи командной оболочки (англ. Shell). Чаще всего в качестве подобной оболочки выступает bash; также используются csh, tsch, zsh и прочие. 

`cat /etc/shells` - посмотреть список доступных в системе шеллов

Команды, используемые в командной строке, относятся к четырем основным типам:

- встроенные команды оболочки (например, read, history; см. список в https://linux.die.net/man/1/bash — раздел Shell Builtin Commands).

- алиасы (например, `alias mv='mv -i'` означает, что команда `mv` при вводе будет преобразовываться в `mv -i`). Их нужно устанавливать в файле `.bashrc`, если требуется использование дольше, чем в течении одной сессии. Уничтожение алиаса происходит при помощи команды `unalias`.
- функции, задаваемые в рамках синтаксиса командной строки. Например, если в файле .bashrc, обрабатываемом при старте командной оболочки, будет задано
```foo() {…}```

то появится возможность пользоваться в командной строке командой foo;

- исполняемые файлы, находящиеся в директориях, перечисленных в переменной окружения PATH.

Проверить, где именно находится исполняемый файл, вызываемый соответствующей командой, можно с помощью утилиты which. Например: `which ls`

## 10. TTY и VTY, разница в подходах к обработке получаемого потока данных.

В линуксе термин vty не используется, вместо него применяется термин псевдотерминал — pts. Это символьные устройства. 

Система из драйвера UART-порта, дисциплины (протокола) и драйвера TTY называется **устройством TTY**, или просто TTY. Естественно, для этого данный процесс должен обладать правами записи в этот файл. Поэтому, когда пользователь входит в систему и подключается к определённому TTY, этот пользователь должен стать владельцем файла, соответствующего этому TTY. Именно это и делает программа login.

Терминалы с электронно-лучевыми трубками сегодня представляют собой не более чем музейные экспонаты, но их дух сохранился в виде **псевдотерминалов**. Эти пары файлов устройств эмулируют текстовый интерфейс терминалов со стороны служб, например, виртуальных консолей, виртуальных терминалов (например, xterm) и служб сетевой регистрации (например, telnet и sh). Драйвер псевдотерминала переносит коды нажатых клавиш и введенный текст между этими двумя устройствами,
скрывая, что физического терминала не существует. Это программная реализация, которая выглядит для присоединенной программы как терминал, но вместо того, чтобы напрямую связываться с «реальным» терминалом, она передает вход и выход другой программе.

Обычно устройства типа tty подключаются к машине напрямую, в то время как pty - это подключения по SSH или telnet.

>Файлы устройств характеризуются двумя номерами: старшим и младшим. Старший номер устройства позволяет ядру определить, к какому драйверу относится файл, а младший номер, как правило, идентифицирует конкретное физическое устройство. Например, старший номер устройства 4 в Linux соответствует драйверу последовательного порта. Таким образом, первый последовательный порт (`/dev/tty0`) будет иметь старший номер 4 и младший номер 0.

`stty` - Печатает или изменяет параметры терминала. 

Последовательным портам персональных компьютеров можно назначать различные адреса ввода-вывода и номера запросов на прерывание (IRQ). Параметры портов считываются из BIOS при включении питания компьютера. Менять их приходится, только
если какое-нибудь “вредное” устройство соглашается работать лишь при выделении ему ресурсов, обычно закрепленных за одним из последовательных портов. К сожалению, драйвер последовательных портов не всегда способен распознать подобные изменения конфигурации без посторонней помощи

К счастью, в системе Linux этого можно избежать, воспользовавшись командой `setserial`. При наличии флага `-g` она отображает текущие установки порта.

Задавая параметры порта, нужно сначала указать файл устройства, а затем — последовательность названий параметров и их значений. К примеру, команда `setserial /dev/ttyS1 port 0х02f8 irq 3` устанавливает адрес ввода-вывода и номер прерывания для порта `ttyS1`.

Расположение конфигурации tty в разных системах различно. В системе Ubuntu для инициализации последовательных портов используется сценарий `/etc/init.d/setserial`. Он считывает параметры для каждого порта из файла `/var/lib/setserial/autoserial.conf`.

Файл `/etc/gettytab` ассоциирует символические имена, например `std.9600`, с профилями конфигурации портов, содержащих такие параметры, как скорость, четность и предложение зарегистрироваться.

## 11. STD In, STD out, STD error (рассказать про управление и работу с потоками).

Одним из основных механзмов ввода/вывода в Unix-подобных (и не только) системах связан с использованием [потоков](https://en.wikipedia.org/wiki/Stream_(computing)) ввода/вывода. Каждому потоку присваивается порядковый номер. Три стандартных потока, открываемых при запуске процесса:

- 0 — stdin (стандартный ввод)
- 1 — stdout (стандартный вывод)
- 2 — stderr (стандартный вывод ошибок)

Подробнее — см. https://en.wikipedia.org/wiki/Standard_streams и https://www.digitalocean.com/community/tutorials/an-introduction-to-linux-i-o-redirection

## 12. Пользователи и группы в ОС Linux.

В UNIX-подобных ОС основными сущностями, определяющими распределение прав доступа и ограничений, являются пользователи и группы.

Аккаунты пользователей и их основные атрибуты содержатся в `/etc/passwd`, хеши паролей — в `/etc/shadow`. Подобное разделение необходимо, т. к. атрибуты пользователей должны быть доступны для чтения всем учетным записям в системе, а хеши паролей — только приложениям, осуществляющим аутентификацию и запущенным с правами суперпользователя.

В процессе регистрации пользователя система обращается к файлу `/etc/passwd` в поисках идентификатора пользователя и его домашнего каталога. Каждая строка файла описывает одного пользователя и содержит семь следующих полей, разделенных двоеточиями:

- регистрационное имя;
- шифрованный пароль или “заполнитель” пароля;
- идентификатор пользователя;
- идентификатор группы по умолчанию;
- поле GECOS (полное имя, номер офиса, рабочий и домашний телефоны);
- домашний каталог;
- регистрационная оболочка

Современные системы помещают в файл `/etc/passwd` некий заполнитель для зашифрованного пароля, а затем во время первой регистрации предлагают пользователю ввести реальный пароль. Они также поддерживают ряд схем шифрования (помимо стандартного UNIX-алгоритма crypt). Если вы решили не пользоваться системными средствами добавления пользователей и для создания новой учетной записи вручную редактировать файл `/etc/passwd` поставьте звездочку (*) или
символ “х” в поле зашифрованного пароля. Звездочка воспрепятствует несанкционированному использованию учетной записи до установки реального пароля. Никогда не оставляйте это поле пустым, иначе безопасность системы окажется под серьезной угрозой, поскольку для доступа к такой учетной записи пароль не требуется.

Файл скрытых (теневых) паролей доступен для чтения только суперпользователю и предназначен для хранения зашифрованных паролей подальше от любопытных глаз. Зашифрованные пароли имеют постоянную длину (34 символа в случае MD5) независимо от длины исходного пароля. Пароли шифруются с добавлением случайной соли, чтобы одному исходному паролю соответствовало много зашифрованных форм. Таким образом, факт выбора пользователями одинаковых паролей не может быть выявлен путем просмотра зашифрованных паролей.

Пароли MD5 легко распознать, так как они всегда начинаются с последовательности `$1$` или `$md5$`. Пароли Blowfish начинаются с последовательности `$2а$`, а пароли SHA256 — с последовательности `$5$`

**Идентификатор пользователя (UID)** — это, как правило, 32-битовое целое число без знака, которое идентифицирует пользователя в системе. Регистрационные имена используются для удобства пользователей, а идентификаторы UID применяются (внутренне) программными средствами и файловой системой. Но для обеспечения совместимости со старыми системами рекомендуем, чтобы значение самого старшего идентификатора по возможности не превышало 32767 (наибольшее 16-битовое целое число со знаком).

В большинстве систем есть также псевдопользователи `bin` и `daemon`, что позволяет им быть владельцами команд и файлов конфигурации. Как правило, такие псевдоимена помещаются в начало файла `/etc/passwd`, и им назначаются маленькие идентификаторы (<500), а также фиктивный интерпретатор команд (например, `/bin/false`).

Предусмотрен еще один специальный идентификатор, именуемый псевдопользователем (**nobody**), которому обычно назначаются такие значения, как -1 или -2 (эти значения, представленные в виде целых без знака для установки поля UID, оказываются максимально большими идентификаторами). Регистрационное имя псевдопользователя используется в случае, если суперпользователь на одном компьютере пытается получить доступ к файлам, смонтированным (посредством системы NFS) с другого компьютера, который не доверяет первому

Как и идентификатор пользователя (UID), **идентификатор группы (GID)** является 32-битовым целым числом. Идентификатор 0 зарезервирован для группы с именем root

Несмотря на наличие поля пароля (с помощью которого пользователи могут присоединиться к группе, выполнив команду `newgrp`) пароль задается редко. Пароль можно установить с помощью команды `gpasswd`; а его зашифрованная форма сохранится в файле `/etc/gshadow`. Групповые пароли используютсядовольно редко.

Если в файле `/etc/passwd` пользователь объявлен членом определенной группы, а в файле `/etc/group` — нет, пользователь все равно включается в группу. На этапе регистрации членство в группе проверяется по обоим файлам, но лучше все же согласовывать их содержимое.

С помощью команды `chfn` можно изменять информацию, содержащуюся в поле GECOS.

>Каждый пользователь может входить в произвольное количество групп; каждая группа может содержать произвольное количество пользователей.

Конфигурация групп находится в `/etc/group`.

Добавлять пользователей можно с помощью команд `adduser` и `useradd`, при этом первая команда является оберткой над второй. Аналогичная команда есть и для менеджмента групп.
