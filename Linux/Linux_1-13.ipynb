## 2. Ядра и их версии. Ванильные ядра.

Ядро Linux разрабатывается под эгидой Linux Foundation.

Общее руководство осуществляют Линус Торвальдс и Грег Кроа-Хартман. В рамках отдельных подсистем ядра решения о принятии тех или иных изменений принимают мейнтейнеры.

Релизы ядра доступны на https://www.kernel.org/. Рекомендуется скачать архив с последней версией ядра, распаковать, ознакомиться со структурой.

Цикл разработки ядра состоит из трех фаз:

- merge window: в master-ветку ядра принимаются изменения, подготовленные разработчиками и одобренные мейнтейнерами (см. файл MAINTAINERS в архиве ядра). Типичная длительность — 2 недели.

- заморозка изменений, с данного момента в текущую версию ядра принимаются только фиксы багов. выпуск релиз-кандидатов. Обычно их 6 — 8, типичное наименование - <следующая версия>-rc<номер релиз-кандидата> (например, 5.5-rc4). Частота выпуска релиз-кандидатов — 1 неделя. В момент, когда руководитель, выпускающий данную версию (Линус или Грег) решает, что в текущем релиз-кандидате багов мало, происходит релиз.

- релиз очередной версии ядра (сопровождающийся открытием merge window для следующей версии). По мере нахождения багов выпускаются исправленные версии — например, 5.4.1, 5.4.2 и т. п. Длительность цикла поддержки большинства ядер — 3 месяца; периодически выпускаются LTS-версии (long term support) ядра, поддерживаемые 2 года.

Изначально смена первого числа версии ядра (1.0, 2.0) была приурочена к глобальной переработке ядра; второе число было четным для стабильных веток (2.0, 2.2, 2.4) и нечетной для активно разрабатываемой (2.1, 2.3, 2.5). Начиная с ядер ветки 2.6 в 2004 году цикл разработки пришел к современному виду: очередная версия получала номер 2.6.X. Вместо версии 2.6.40 было выпущено ядро 3.0 без каких-либо кардинальных изменений; очередные версии получили номера 3.Х. Выпуск версий 4.0 и 5.0 также не был связан с крупными изменениями ядра.

Ядра, выпускаемые linux foundation и размещенные на kernel.org, называются «ванильными». Многие дистрибутивы собирают видоизмененные ядра с дополнительными драйверами, бекпортированным из более новых версий функционалом и т. п. Например, в ядре 3.10.0-1062.7.1 из дистрибутива RHEL 7 присутствует функционал из версий ядра вплоть до 4.18.

## 5. Список поддерживаемых файловых систем ОС Linux. LVM.

Журналирование ФС: https://en.wikipedia.org/wiki/Journaling_file_system

По всем нижеперечисленным ФС прочитать статьи в англоязычной википедии.

- семейство ext: ext2/ext3/ext4

ext2 — нежурналируемая файловая система. Ограничения: размер файла 2 Тб, размер тома 32 Тб. Неэффективная работа с директориями с большим количеством файлов.

Ext3 — журналируемая ФС с поддержкой htree для более эффективной работы с директориями, содержащими большое кол-во файлов.

ext4 — дальнейшее развитие. Ограничение размера ФС — 1 экзобайт, размера файла — 32 Тб. Поддержка преаллокации, отложенной аллокации, extent-ов — см. https://en.wikipedia.org/wiki/Extent_(file_systems).

- xfs — журналируемая ФС с акцентом на параллелизм и поддержку эффективной работы с большими файлами. Ограничение размера — 8 экзобайт.

- btrfs позиционируется как ФС нового поколения. Поддерживает copy on write, прозрачное сжатие, снепшоты, объединение устройств в raid-массивы. Максимальный размер — 8 экзобайт.

- множество файловых систем для использования на определенных типах устройств, разработанных для специальных задач и т. п. (f2fs, overlayfs, тысячи их).

LVM: https://access.redhat.com/documentation/en-us/red_hat_enterprise_linux/8/html/configuring_and_managing_logical_volumes/index

## 6. Стандарт POSIX и его влияние на ОС Linux. FHS. 

https://en.wikipedia.org/wiki/POSIX Стандарт POSIX предназначен для обеспечения переносимости исходных кодов приложений между UNIX-подобными ОС. Поддержка стандарта POSIX в линуксе позволила портировать многие существующие приложения с минимальными изменениями, что способствовало его распространению.

https://en.wikipedia.org/wiki/Filesystem_Hierarchy_Standard

В отличие от ОС Windows, структура директорий в которой привязана к присутствующим в системе разделам (c:, d:) и т. п., в UNIX-подобных ОС структура директорий не зависит от конкретной конфигурации устройств хранения: существует единая иерархия дерева

директорий, а разнообразные хранилища (диски, сетевые ФС, псевдо-ФС и т. п.) могут монтироваться к произвольным путям в вышеупомянутом дереве.

## 7. Работа с устройствами как с файлами. Процесс подключения устройств к системе FHS.

Взаимодействие с многими аппаратными устройствами в UNIX-подобных ОС осуществляется через специальные файлы, размещенные на псевдо-ФС devtmpfs.

Примерами подобных устройств являются HDD/SSD (типичные названия устройств sdX / nvmeX), терминалы (ttyX), псевдотерминалы (ptsX), специальные устройства null (принимает на запись произвольные данные и забывает про них; при чтении отдает EOF), zero (при чтении отдает произвольное количество нулей), random и urandom (источники произвольного количества случайных чисел с гарантией использования достаточного кол-ва энтропии и без гарантии; см. https://en.wikipedia.org/wiki/Entropy_(computing)).

Псевдо-файловые устройства делятся на блочные и символьные. Операции ввода/вывода с символьными устройствами осуществляются побайтово, с блочными — поблочно (размер блока зависит от свойств конкрентного устройства).

Создание файла устройства из командной строки осуществляется с помощью команды mknod, из кода программы — с помощью библиотечной функции mknod (прочитать http://man7.org/linux/man-pages/man1/mknod.1.html и https://linux.die.net/man/3/mknod).

Процесс «подключения устройств» к дереву директорий называется монтированием. Участниками операции монтирования являются файловая система, находящаяся на блочном устройстве, удаленном хосте, в файле и т. п., а также точка монтирования — уже существующая директория, внутри которой будет доступно содержимое монтируемой ФС.

Аргумент -t следует указывать, если автоматическое распознавание типа ФС по тем или иным причинам невозможно или не приводит к желаемому результату.

Аргумент -o позволяет указать конкретные опции монтирования. Набор опций, поддерживаемых той или иной ФС, приведен на соответствующей странице руководства — например, man ext4, раздел MOUNT OPTIONS.

Примеры:

mount /dev/sdb1 /mnt — монтирование файловой системы, находящейся на разделе /dev/sdb1 в директорию /mnt;

mount /dev/sdb1 -t ext4 -o noatime /mnt — то же самое, но с указанием типа ФС и опции noatime; mount 192.168.1.1:/path/to/websites /var/www — монтирование сетевой файловой системы (NFS), находящейся на хосте 192.168.1.1 по пути /path/to/websites в директорию /var/www.

mount -o loop /home/vasya/mycoolgame.iso /mnt — монтирование образа ФС, находящегося в файле /home/vasya/mycoolgame.iso в директорию /mnt.

При перезагрузке монтирование, выполненное вышеуказанным способом, не сохраняется. Список ФС, подлежащих монтированию при загрузке, приводится в /etc/fstab — см. http://man7.org/linux/man-pages/man5/fstab.5.html

## 8. CLI ОС Linux. Основные команды навигации.

В качестве командной оболочки в linux чаще всего используется bash. Примеры навигации в bash:

pwd — вывести текущую директорию.

cd <путь> - перейти в указанную директорию

Основные разновидности путей — относительный и абсолютный. Абсолютный путь начинается с / - корневой директории — и не зависит от того, в какой директории он применяется. Относительный путь не начинается с / и применяется относительно текущей директории.

cd (без аргументов) — перейти в домашнюю директорию текущего пользователя.

Тильда (~) в пути обозначает домашнюю директорию текущего пользователя; ~<логин> - домашнюю директорию пользователя, логин которого приведен; например, ~vasya — домашняя директория пользователя vasya.

ls <путь> выводит список файлов и директорий, находящихся по указанному пути.

Основные опции:

-l — отображение свойств файлов и директорий;

-d — отображение свойств указанной директории, а не ее содержимого;

-t — сортировка отображаемых файлов по времени изменения от новых к старым;

-a — включение отображения скрытых файлов и директорий (скрытыми являются файлы и директории, название которых начинается с точки - '.').

## 9. Shells ОС Linux. Процесс обработки введенных команд.

Работа в командной строке осуществляется при помощи командной оболочки (англ. Shell). Чаще всего в качестве подобной оболочки выступает bash; также используются csh, tsch, zsh и прочие.

Команды, используемые в командной строке, относятся к четырем основным типам:

- встроенные команды оболочки (например, read, history; см. список в https://linux.die.net/man/1/bash — раздел Shell Builtin Commands).

- алиасы (например, alias mv='mv -i' означает, что команда mv при вводе будет преобразовываться в mv -i)
- функции, задаваемые в рамках синтаксиса командной строки. Например, если в файле .bashrc, обрабатываемом при старте командной оболочки, будет задано
```foo() {…}```

то появится возможность пользоваться в командной строке командой foo;

- исполняемые файлы, находящиеся в директориях, перечисленных в переменной окружения PATH.

Проверить, где именно находится исполняемый файл, вызываемый соответствующей командой, можно с помощью утилиты which. Например: `which ls`

## 11. STD In, STD out, STD error (рассказать про управление и работу с потоками).

Одним из основных механзмов ввода/вывода в Unix-подобных (и не только) системах связан с использованием потоков ввода/вывода (см. https://en.wikipedia.org/wiki/Stream_(computing)). Каждому потоку присваивается порядковый номер. Три стандартных потока, открываемых при запуске процесса:

0 — stdin (стандартный ввод); 1 — stdout (стандартный вывод);

2 — stderr (стандартный вывод ошибок).

Подробнее — см. https://en.wikipedia.org/wiki/Standard_streams и https://www.digitalocean.com/community/tutorials/an-introduction-to-linux-i-o-redirection

## 12. Пользователи и группы в ОС Linux.

В UNIX-подобных ОС основными сущностями, определяющими распределение прав доступа и ограничений, являются пользователи и группы.

Аккаунты пользователей и их основные атрибуты содержатся в /etc/passwd, хеши паролей — в /etc/shadow. Подобное разделение необходимо, т. к. атрибуты пользователей должны быть доступны для чтения всем учетным записям в системе, а хеши паролей — только приложениям, осуществляющим аутентификацию и запущенным с правами суперпользователя.

Атрибуты пользователя — uid, gid основной группы, логин, домашняя директория, командная оболочка.

https://linux.die.net/man/5/passwd

Каждый пользователь может входить в произвольное количество групп; каждая группа может содержать произвольное количество пользователей.

Конфигурация групп находится в /etc/group.

https://linux.die.net/man/5/group
